```python
import pandas as pd
```

```python
df_initial = pd.DataFrame()
df_accident = pd.DataFrame()
df_coordinates = pd.DataFrame()
df_road = pd.DataFrame()
df_weather = pd.DataFrame()
```

```python
df_initial = pd.read_csv("Accident_Information.csv")
```

```python
df_accident = df_initial[['Accident_Severity','Light_Conditions', 'Number_of_Vehicles', 'Time', 'Urban_or_Rural_Area', 'Year']]
```

```python
df_accident.insert(0, 'Class', "Accident")
```

```python
df_accident
```

```python
df_accident.insert(1, 'Accident_Individuals', range(1, 1 + len(df_accident)))
df_accident['Accident_Individuals'] = 'accident_' + df_accident['Accident_Individuals'].astype(str)
```

```python
df_accident
```

```python
df_accident.insert(2, 'Coordinate_Individuals', range(1, 1 + len(df_accident)))
df_accident['Coordinate_Individuals'] = 'coordinate_' + df_accident['Coordinate_Individuals'].astype(str)

df_accident.insert(3, 'Weather_Individuals', range(1, 1 + len(df_accident)))
df_accident['Weather_Individuals'] = 'weather_' + df_accident['Weather_Individuals'].astype(str)

df_accident.insert(4, 'Road_Individuals', range(1, 1 + len(df_accident)))
df_accident['Road_Individuals'] = 'road_' + df_accident['Road_Individuals'].astype(str)
```

```python
df_accident
```

```python
df_accident.loc[df_accident['Accident_Severity'] == "Slight", 'Accident_Severity'] = 1
df_accident.loc[df_accident['Accident_Severity'] == "Serious", 'Accident_Severity'] = 2
df_accident.loc[df_accident['Accident_Severity'] == "Fatal", 'Accident_Severity'] = 3
```

```python
df_accident.loc[df_accident['Light_Conditions'] == "Daylight", 'Light_Conditions'] = "Day"
df_accident.loc[df_accident['Light_Conditions'] == "Darkness - lights lit", 'Light_Conditions'] = "Night with visibility"
df_accident.loc[df_accident['Light_Conditions'] == "Darkness - no lighting", 'Light_Conditions'] = "Dark"
df_accident.loc[df_accident['Light_Conditions'] == "Darkness - lighting unknown", 'Light_Conditions'] = "Dark"
df_accident.loc[df_accident['Light_Conditions'] == "Darkness - lights unlit", 'Light_Conditions'] = "Night with low visibility"
```

```python
df_accident.to_csv("accident.csv")
```

```python
df_coordinates = df_initial[['Latitude','Longitude', 'Local_Authority_(District)']]
df_coordinates.insert(0, 'Class', "Coordinate")

df_coordinates.insert(1, 'Coordinate_Individuals', range(1, 1 + len(df_coordinates)))
df_coordinates['Coordinate_Individuals'] = 'coordinate_' + df_coordinates['Coordinate_Individuals'].astype(str)

df_coordinates.insert(2, 'Accident_Individuals', range(1, 1 + len(df_coordinates)))
df_coordinates['Accident_Individuals'] = 'accident_' + df_coordinates['Accident_Individuals'].astype(str)
```

```python
df_road = df_initial[['Road_Surface_Conditions', 'Road_Type', 'Speed_limit']]
df_road.insert(0, 'Class', "Road")

df_road.insert(1, 'Road_Individuals', range(1, 1 + len(df_road)))
df_road['Road_Individuals'] = 'road_' + df_road['Road_Individuals'].astype(str)

df_road.insert(2, 'Accident_Individuals', range(1, 1 + len(df_road)))
df_road['Accident_Individuals'] = 'accident_' + df_road['Accident_Individuals'].astype(str)
```

```python
df_weather = df_initial[['Weather_Conditions']]
df_weather.insert(0, 'Class', "Weather")

df_weather.insert(1, 'Weather_Individuals', range(1, 1 + len(df_weather)))
df_weather['Weather_Individuals'] = 'weather_' + df_weather['Weather_Individuals'].astype(str)

df_weather.insert(2, 'Accident_Individuals', range(1, 1 + len(df_weather)))
df_weather['Accident_Individuals'] = 'accident_' + df_weather['Accident_Individuals'].astype(str)

df_weather.loc[df_weather['Weather_Conditions'] == "Fine no high winds", 'Weather_Conditions'] = "Normal"
df_weather.loc[df_weather['Weather_Conditions'] == "Raining no high winds", 'Weather_Conditions'] = "Rainy"
df_weather.loc[df_weather['Weather_Conditions'] == "Other", 'Weather_Conditions'] = "Other"
df_weather.loc[df_weather['Weather_Conditions'] == "Unknown", 'Weather_Conditions'] = "Other"
df_weather.loc[df_weather['Weather_Conditions'] == "Fine + high winds", 'Weather_Conditions'] = "Stormy"
```
